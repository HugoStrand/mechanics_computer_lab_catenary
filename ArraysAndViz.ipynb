## Array Slicing

Syntax `start:end:step`. Negative indices count from the end with `-1` being the last element.

```python
x[0:3]
```

```python
x[1:4]
```

```python
x[4:0:-1]
```

```python
x[::-1]
```

## Element-wise operations

# Visualisation

## Matplotlib

The standard plotting module in Python, using Numpy arrays. With comprehensive documentation, see https://matplotlib.org/.

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 3, num=100)
y = 1 + x**3/8
```

```python
plt.plot(x, y)
```

## Limits, labels and legends

```python
plt.plot(x, y, label='$y(x) = 1+x^3/8$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(loc='best')
plt.ylim([0, 5])
plt.xlim([-1, 4])
```

## Styles

- Line styles: `solid`, `dotted`, `dashed`, `dashdot`, ...

  https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
  
- Markers: `o`, `s`, `x`, `+`, etc.

  https://matplotlib.org/stable/api/markers_api.html

```python
linestyles = ['solid', 'dotted', 'dashed', 'dashdot']
markers = ['o', 's', 'x', '+']

for idx, (linestyle, marker) in enumerate(zip(linestyles, markers)):
    x = np.arange(10)
    y = idx + np.random.random(len(x))
    plt.plot(x, y, linestyle=linestyle, marker=marker)
```
